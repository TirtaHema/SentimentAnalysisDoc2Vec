Using credential from twitter API

```python
auth = tw.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

Get data from twitter

```python
search_words = "komodo"
date_since = "2020-10-25"
```

```python
tweets = tw.Cursor(api.search,
              q=search_words,
              lang="id",
              since=date_since,
              tweet_mode="extended").items(2000)
```

```python
result =[]
for tweet in tweets:
  if 'retweeted_status' in tweet._json:
    temp = [tweet.user.screen_name,tweet._json['retweeted_status']['full_text']]
  else:
    temp = [tweet.user.screen_name,tweet.full_text]
  result.append(temp)
```

```python
len(result)
```

```python
tweet_text = pd.DataFrame(data=result, 
                    columns=['user', "tweets"])
tweet_text
```
